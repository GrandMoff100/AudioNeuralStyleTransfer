```python
import io
import numpy as np
import pickle
from pydub import AudioSegment
from scipy.io import wavfile
from scipy.signal import stft
from tempfile import mktemp


def convert_song(file: str):
    mp3_audio = AudioSegment.from_file(file, format="mp3")
    temp = io.BytesIO()
    mp3_audio.export(temp, format="wav")
    FS, data = wavfile.read(temp)
    _, _, spectrogram = stft(data, fs=FS)
    del data, FS, mp3_audio, temp
    target_path = f"/content/drive/Shareddrives/Beural Style Transfer/Spectrograms/{file.split('/')[-1].split('.')[0] + '.dat'}"
    print("Saving to", target_path, "\n")
    with open(target_path, "wb") as file:
        pickle.dump(spectrogram, file)
```

```python
import glob
import warnings

warnings.filterwarnings("ignore")

songs = glob.glob("/content/drive/Shareddrives/Beural Style Transfer/NewMusic/*/*.mp3")
```

```python
for i, song in enumerate(songs[100:200]):
    print(i, song)
    convert_song(song)
```
